## Make API call

```python
# Import the requests library.
import requests
import pandas as pd
import numpy
from datetime import datetime
```

```python
# Starting URL for Weather API Call.
url ="https://www.ncei.noaa.gov/access/past-weather/USW00012923/data.json"
data = "data"
```

```python
# writing the query
query_url = url
```

```python
# checking the API response.
weather_data = requests.get(query_url)
weather_data
```

```python
weather_data.status_code
```

```python
# Get the text of the 'Get' request.
weather_data.text
```

```python
# Get the JSON text of the 'Get' request.
weather_data = weather_data.json()
weather_data
```

```python
weather_data = weather_data['data']
weather_data
```

## Read the json data into a dataframe.

```python
# Convert the array of dictionaries to a Pandas DataFrame.
weather_data_df = pd.DataFrame(weather_data)
weather_data_df.head(10)
```

## Explore and Clean the dataframe

```python
# swap rows and columns
weather_data_df = weather_data_df.transpose()
weather_data_df
```

```python
# move date index to a column.
weather_data_df = weather_data_df.reset_index()
weather_data_df.head(2)
```

```python
weather_data_df = weather_data_df.drop(columns=["SNOW", "SNWD"])
weather_data_df.head(2)
```

```python
# rename the date column
weather_data_df.rename(columns={"index":"date", "TMAX":"max_temp", "TMIN":"min_temp", "PRCP":"precipitation", "TAVG":"avg_temp"}, inplace = True)
weather_data_df.columns
```

```python
weather_data_df.describe()
```

```python
weather_data_df.shape
```

```python
weather_data_df.info()
```

```python
weather_data_df.dropna(inplace=True)

weather_data_df.shape
```

## Convert necessary datatypes

```python
# convert date object column into datetime.
weather_data_df["date"] = weather_data_df["date"].apply(lambda x:datetime.strptime(x,"%Y-%m-%d"))
```

```python
# convert 'max_temp', 'min_temp', 'avg_temp' from objects to integers.
cols = ['max_temp', 'min_temp', 'avg_temp']
weather_data_df[cols] = weather_data_df[cols].apply(pd.to_numeric, errors='coerce', axis=1)
```

```python
# convert 'precipitation' column from object to float
weather_data_df['precipitation'] = weather_data_df['precipitation'].astype(str).astype(float)
```

```python
# check the data types again.
weather_data_df.dtypes
```

```python
# sort df in descending order based on the date column and rearange the index sequence.
weather_data_df = weather_data_df.sort_values(by="date",ascending=False, ignore_index=True)
weather_data_df.head(2)
```

# 5 days rolling window

```python
# calculating simple moving average
# using .rolling(window).mean() ,
# with window size = 5
weather_data_df['5_days_max_temp'] = weather_data_df['max_temp'].rolling(5).mean()
weather_data_df['5_days_min_temp'] = weather_data_df['min_temp'].rolling(5).mean()
weather_data_df['5_days_avg_temp'] = weather_data_df['avg_temp'].rolling(5).mean()
weather_data_df['5_days_avg_temp'] = weather_data_df['precipitation'].rolling(5).mean()

weather_data_df.head(5)
```
